### Data Cleaning

```python
# Filling missing values
loan_data['Gender'].fillna('Male' , inplace=True)
loan_data['Married'].fillna('No' , inplace=True)
loan_data['Credit_History'].fillna(1.0 , inplace=True)
loan_data['Self_Employed'].fillna('No' , inplace=True)
loan_data['Loan_Amount_Term'].fillna(loan_data['Loan_Amount_Term'].median() , inplace=True)
loan_data['LoanAmount'].fillna(loan_data['LoanAmount'].median() , inplace=True)
loan_data['Dependents'].fillna('0' , inplace = True)
```

```python
loan_data['Dependents'] = loan_data['Dependents'].str.replace('+' , '' , regex = False)
loan_data['Dependents'] = pd.to_numeric(loan_data['Dependents'])
loan_data['Credit_History'] = pd.to_numeric(loan_data['Credit_History'])
```

```python
loan_data['Loan_Status'] = [1.0 if c == 'Y' else 0.0 for c in loan_data['Loan_Status']]
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from imblearn.over_sampling import SMOTE
```

```python
loan_data.columns
```

```python
df = loan_data.drop(columns=['Loan_ID'])
```

```python
df['total_income'] = df['ApplicantIncome'] + df['CoapplicantIncome']
```

```python
# df['Income_Loan_ratio'] = df['LoanAmount'] / (df['total_income'].replace(0,1))
```

```python
df['Income_Per_Dependent'] = df['total_income'] / (df['Dependents'].replace(0, 1))
df['Loan_Term_Years'] = df['Loan_Amount_Term'] / 12
df['Monthly_Installment'] = df['LoanAmount'] / df['Loan_Amount_Term']
df['Installment_Income_Ratio'] = df['Monthly_Installment'] / df['total_income']
df['ApplicantIncome_Log'] = np.log1p(df['ApplicantIncome'])
df['LoanAmount_Log'] = np.log1p(df['LoanAmount'])
```

```python
df
```

```python
X = df.drop(columns=['Loan_Status'])
y = df['Loan_Status']
```

```python
X.isnull().sum()
```

```python
X.info()
```

```python
num_features = X.select_dtypes(exclude='object').columns
```

```python

preprocessor = ColumnTransformer(transformers = [
    ('Ordinals',OrdinalEncoder(categories=[['Not Graduate','Graduate'],['Rural','Semiurban','Urban']], dtype=np.int64) , ['Education', 'Property_Area']),
    ('Binary', OneHotEncoder(), ['Self_Employed', 'Married', 'Gender'] ),
    ('StandardScaler' , StandardScaler() , num_features)
], remainder='passthrough')

```

```python
X = preprocessor.fit_transform(X)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , random_state=45 , test_size=0.2)
```

```python
# oversampling 
smote = SMOTE(random_state=45)
X_train_res , y_train_res = smote.fit_resample(X_train , y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
```

```python
models = {
    'Decision Tree' : DecisionTreeClassifier(),
    'Rnadom Forest' : RandomForestClassifier(),
    'Logistic Regression' : LogisticRegression(),
    'XGB': XGBClassifier(),
    'LGBM': LGBMClassifier(),
}
```

```python
for i in range(len(models)):
    model = list(models.values())[i]
    model.fit(X_train , y_train)
    
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    print('*'*32)
    print(list(models.keys())[i])
    print('*'*32)
    print('Test Score')
    print('-'*32)
    print(classification_report(y_true= y_test , y_pred=y_test_pred))
    
    
    
    
    
```
